### Agent with Memory


```python
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver
```

```python
import os
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
```

```python
agent=create_agent( 
    model="groq:llama-3.1-8b-instant",
    checkpointer=InMemorySaver() # for memory
)
```

```python
### Run with Thread id -1
config={
    "configurable": {"thread_id": "test1"}
}
```

```python
from langchain_core.messages import HumanMessage,SystemMessage
response=agent.invoke({"messages":[HumanMessage(content="Hi my name is Hardik!")]},config=config)
response
```

```python
response=agent.invoke({"messages":[HumanMessage(content="Hi What's my name ?")]},config=config)
response["messages"][-1]
```

```python
### Run with Thread id -2
config1={
     "configurable": {"thread_id": "test2"}
}
response=agent.invoke({"messages":[HumanMessage(content="Hi What's my name ?")]},config=config1)
response["messages"][-1]
#  Shows no information as different thread id we are using and we havn't provide
#  any context to it
```
